### Context
Given $h\colon \mathbb{R}^n\to \mathbb{R}^m$ be a smooth function and $C\subset \mathbb{R}^{m}$ be a convex set. The aim of this project is to study the problem of finding an approximate projection for the set $h^{-1}(C)$ from the theoretical and numerical point of view. Hence, given $x\in \mathbb{R}^{n}$, we aim to solve the problem
$$
    \min_{y\in h^{-1}(C)}\frac{1}{2}\Vert x-y\Vert^2 \Leftrightarrow  \min_{h(y)\in C}\frac{1}{2}\Vert x-y\Vert^2
$$

A prominent example is when the functions $h_i$ are weakly convex.

### Tasks

1. To give conditions on $h$ in such a way that $h^{-1}(C)$ is uniformly prox-regular.(Emilio-Juan)
2. To provide numerical algorithms to solve (approximately) the projection problem. 

A first attempt would be to consider the composite problem:
$$
\min_{y\in \mathcal{H}}\frac{1}{2}\Vert x-y\Vert^2+\frac{1}{2\gamma}d_{C}^2(h(y)),
$$
when $\gamma$ goes to $0$. In fact, if $y_\gamma$ solves the composite problem then one has
$$
    \|x-y_\gamma\|^2 + \frac{1}{\gamma}d_{C}^2(h(y_\gamma))\leq d_{h^{-1}(C)}^2(x)
$$
So, the cluster points of $(y_\gamma)$ are minimizers of $\min_{y\in h^{-1}(C)}\frac{1}{2}\Vert x-y\Vert^2$, then if this equation has only one solution, then $(y_\gamma)$ converges to it since it is bounded. {\color{red} En general, ¿$(y_\gamma)$ converge fuerte en dimensión finita?}

### Examples

As starter, we will consider the following two examples:


1. $C=\mathbb{R}_{-}$ and $h\colon \mathbb{R}^n\to \mathbb{R}$ is defined by
$$
\frac{1}{2}x^t Ax+b^tx+c,
$$
where $A$ is symmetric.


2. $C=[-1,1]$ and $h\colon \mathbb{R}^n\to \mathbb{R}$ is defined by
$$
\frac{1}{2}x^t Ax+b^tx+c,
$$
where $A$ is symmetric.

### Theorems

**Theorem:** An $n \times n$ matrix is orthogonally diagonalizable if and only if is a symmetric matrix

**Theorem(QR Decomposition):**  Any real square matrix may be decompose as $A = QR$ where $Q$ Q is an orthogonal matrix and $R$ is an upper triangular matrix 

There are several methods for actually computing the QR decomposition, such as the Gram–Schmidt process, Householder transformations, or Givens rotations. 

**Obs:** The skew-symmetric decomposition of a matrix does not contribute to the quadratic form, that is why we consider $A$ a symmetric matrix without loss of generality, this is, if $Q_{A}(x)$ is a quadratic form and $A = -A^{T} \iff Q_{A}(x) = 0$

**Theorem: (Rayleigh-Quotient)** For any given symmetric matrix $A \in \mathbb{R}^{n \times n}$
$$\max_{x \in \mathbb{R}^n : x \neq 0} \frac{\langle Ax, x \rangle}{||x||^{2}} = \lambda_{max}(A)$$
$$\min_{x \in \mathbb{R}^n : x \neq 0} \frac{\langle Ax, x \rangle}{||x||^{2}} = \lambda_{min}(A)$$

**Obs:** $D^{2}h(x) = A \ \forall x \in \mathbb{R^n}$, in particular, 
$$\lambda_{min}(D^{2}h(x))||\varphi||^{2} \leq \langle D^{2}h(x)\varphi, \varphi \rangle \ \leq \lambda_{max} (D^{2}h(x)) ||\varphi||^{2} \ \forall x \in \mathbb{R}^{n} \ \forall \varphi \in \mathbb{R}^{n} $$


**From J. P Vial Strong and Weak Convexity of Sets and Functions(1983)**

**Def:** Let $f: D \rightarrow R$ be a real-valued function from a convex subset $D$ of $E^n . f$ is said to be $\rho$-convex if there exists some real number $\rho$ such that $\forall x_1, x_2 \in D$, $\forall \lambda \in[0,1]$
$$
f\left(x_1+(1-\lambda) x_2\right) \leqslant \lambda f\left(x_1\right)+(1-\lambda) f\left(x_2\right)-\rho \lambda(1-\lambda)\left\|x_1-x_2\right\|^2 .
$$

If $\rho>0$, the function is said to be strongly convex.
If $\rho=0$, the function is convex.
If $\rho<0$, the function is said to be weakly convex.

**Prop:** $f: E^n \rightarrow E$ is $\rho$-convex if and only if there exists a convex function $h: E^n \rightarrow E$ such that $f(x)=h(x)+\rho\|x\|^2$.

**Theorem**: A twice differentiable function is $\rho$-convex if and only if there exists some real $\rho$ such that for any $x \in E^n$ and $v \in E^n$
$$
\left\langle\nabla^2 f(x) v, v\right\rangle \geqslant 2 \rho\|v\|^2 .
$$

**Obs:** In particular, a quadratic form is $\rho$-convex with respect to $\rho=e / 2$, where $e$ is the smallest eigenvalue of the Hessian matrix.

```python
# To define h: R^n->R given the parameters
def h(x, A, b, c):
    """
    **Quadratic form function.**

    Parameters:
        x (numpy.ndarray): Input vector.
        A (numpy.ndarray): Symmetric matrix.
        b (numpy.ndarray): Vector.
        c (float): Scalar.

    Returns:
        float: Value of the quadratic form function at point x.
    """
    return 0.5 * np.dot(x.T, np.dot(A, x)) + np.dot(b.T, x) + c

# Create random n x n symmetric matrix with eigen-values (\lambda_1, ..., \lambda_n)
def symmetric_matrix(eigenvalues):
    """
    **Generate a random symmetric matrix with given eigenvalues.**

    Parameters:
        eigenvalues (list): List of eigenvalues.

    Returns:
        tuple: Tuple containing the generated symmetric matrix and the orthogonal matrix used for generation.
    """
    n = len(eigenvalues)
    
    D = np.diag(eigenvalues)
    random_matrix = np.random.rand(n, n) 
    Q, _ = np.linalg.qr(random_matrix)
    symm_matrix = np.dot(np.dot(Q, D), Q.T)
    
    return symm_matrix, Q

# Example:
eigen_value = (1, 2, 3)
matriz_resultante, Q = symmetric_matrix(eigen_value)

print("Valores propios originales:", eigen_value)
print("Matriz simétrica generada:")
print(matriz_resultante)
print("Matriz ortogonal Q utilizada:")
print(Q)
```

#### Testing solutions with 'trust-constr' method from Scipy [Trust-region constrained method]

We use the trust-region constrained method to solve constrained minimization problems. This method assumes that the problems have the following form:
$$
\begin{aligned}
& \min _{\mathbf{x}} f(\mathbf{x}) \\
& \text { subject to } \\
& \mathbf{b}_l \leq \mathbf{x} \leq \mathbf{b}_u \\
& \mathbf{c}_l \leq L \mathbf{x} \leq \mathbf{c}_u \\
& \mathbf{d}_l \leq \mathbf{g}(\mathbf{x}) \leq \mathbf{d}_u
\end{aligned}
$$
where
- $\mathbf{x} \in \mathbb{R}^n$ is the $\mathrm{n}$-dimensional vector of optimization variables
- $f(\mathbf{x}): \mathbb{R}^n \rightarrow \mathbb{R}$ is the cost function
- $\mathbf{b}_l, \mathbf{b}_u \in \mathbb{R}^n$ are the vectors of lower and upper bounds on the optimization variable $\mathbf{x}$ **(Bounds)**
- $\mathbf{c}_l, \mathbf{c}_u \in \mathbb{R}^l$ are the vectors of lower and upper bounds on the linear term $L \mathbf{x}$, where $L \in \mathbb{R}^{l \times n}$ is a constant matrix of parameters. The compact notation
$$
\mathbf{c}_l \leq L \mathbf{x} \leq \mathbf{c}_u
$$
represents a set of $l$ linear "less than or equal to" constraints. **(Linear Constraint)**
- $\mathbf{d}_l, \mathbf{d}_u \in \mathbb{R}^p$ are the vectors of lower and upper bounds on the nonlinear vector function $\mathbf{g}(\mathbf{x}): \mathbb{R}^n \rightarrow \mathbb{R}^p$. The compact notation
$$
\mathbf{d}_l \leq \mathbf{g}(\mathbf{x}) \leq \mathbf{d}_u 
$$
represents a set of "p" nonlinear "less than or equal to" constraints. **(Nonlinear Constraint)**

All the 'less-than' or 'equal to' inequalities are interpreted element-wise. For example, the notation
$$
\mathbf{b}_l \leq \mathbf{x} \leq \mathbf{b}_u
$$
is actually a shorthand notation for
$$
\begin{array}{r}
b_{l 1} \leq x_1 \leq b_{u 1} \\
b_{l 2} \leq x_2 \leq b_{u 2} \\
\mathbf{b}_l=\left[\begin{array}{l}
b_{l 1} \\
b_{l 2}
\end{array}\right], \quad \mathbf{b}_u=\left[\begin{array}{l}
b_{u 1} \\
b_{u 2}
\end{array}\right], \quad \mathbf{x}=\left[\begin{array}{l}
x_1 \\
x_2
\end{array}\right]
\end{array}
$$
where for simplicity, we assumed $n=2$.

Then, the one-sided less than or equal to inequalities of the following form
$$
\mathbf{x} \leq \mathbf{b}_u
$$
are formally written as follows
$$
\begin{array}{r}
\mathbf{b}_l \leq \mathrm{x} \leq \mathbf{b}_u \\
\mathbf{b}_l=\left[\begin{array}{c}
-\infty \\
-\infty \\
\vdots \\
-\infty
\end{array}\right]
\end{array}
$$

In a similar manner, we can formally write other types of one-sided less than or equal inequalities.

**From Scipy docs**

"Method trust-constr is a trust-region algorithm for constrained optimization. It switches between two implementations depending on the problem definition. It is the most versatile constrained minimization algorithm implemented in SciPy and the most appropriate for large-scale problems."

1. For **equality constrained** problems it is an implementation of Byrd-Omojokun Trust-Region SQP method described in Lalee, Marucha, Jorge Nocedal, and Todd Plantega. 1998. 'On the implementation of an algorithm for large-scale equality constrained optimization' pp.682-706. and in Nocedal, J, and S J Wright. 2006. 'Numerical Optimization.', p. 549.

2. When **inequality constraints** are imposed as well, it switches to the trust-region interior point method described in Byrd, Richard H., Mary E. Hribar, and Jorge Nocedal. 1999. 'An interior point algorithm for large-scale nonlinear programming'. SIAM Journal on Optimization 9.4: 877-900. This interior point algorithm, in turn, solves inequality constraints by introducing slack variables and solving a sequence of equality-constrained barrier problems for progressively smaller values of the barrier parameter. The previously described equality constrained SQP method is used to solve the subproblems with increasing levels of accuracy as the iterate gets closer to a solution.

More references: Conn, A. R., Gould, N. I., & Toint, P. L. Trust region methods. 2000. Siam. pp. 19.

```python
from scipy.optimize import NonlinearConstraint

# Objective function
def objetive_function_aux(y, x):
    """
    **Auxiliary objective function.**

    Parameters:
        y (numpy.ndarray): Input vector.
        x (numpy.ndarray): Input vector.

    Returns:
        float: Value of the auxiliary objective function 0.5*||x-y||^{2} at point x.
    """
    return 0.5 * (np.linalg.norm(x - y) ** 2)

# Parameters
method = 'trust-constr' #For non-linear constraints
x_0 = np.array([2,2])

#Constraint for C = R_{-}
cons_R_minus_h1 = NonlinearConstraint(h_1, -np.inf, 0)
cons_R_minus_h2 = NonlinearConstraint(h_2, -np.inf, 0)

#Constraint for C = [-1,1]
cons_C_h1 = NonlinearConstraint(h_1, -1, 1)
cons_C_h2 = NonlinearConstraint(h_2, -1, 1)
```

```python
# Solving for C = R_{-} and h_1(x) (weakly-convex)
x_given = np.array([-1,8])

def objective_function(y):
    return objetive_function_aux(y,x_given)

sol_h1 = minimize(objective_function, x_0, method= method, constraints=cons_R_minus_h1)
opt_y = sol_h1.x 
print('Projection into [h1(x) < 0] given x =', x_given, 'is:', opt_y)

# Test: If y \in h^{-1}(C) returns True, return False in other case 
print('Does optimal y belong to h^{-1}(C)?:', h_1(opt_y) <= 0)

# Solving for C = R_{-} and h_2(x) (convex-programming)

sol_h2 = minimize(objective_function, x_0, method= method, constraints=cons_R_minus_h2)
opt_y = sol_h2.x 
print('Projection into [h2(x) < 0] given x =', x_given, 'is:', opt_y)

# Test: If y \in h^{-1}(C) returns True, return False in other case 
print('Does optimal y belong to h^{-1}(C)?:', h_2(opt_y) <= 0)
```

#### Functions to perform the test

```python
def generate_tuple_eigenvalues(n):
    """
    **Generate a tuple of eigenvalues with at least one negative value.**

    Parameters:
        n (int): Number of eigenvalues.

    Returns:
        tuple: Tuple of eigenvalues.
    """
    n = max(1, n)
    array = np.random.randint(-10, 11, n - 2)
    array = np.append(array, np.random.randint(-15, 0))
    array = np.append(array, np.random.randint(0, 15))
    np.random.shuffle(array)
    return tuple(array)

def generate_random_parameters(n):
    """
    **Generate random parameters for a quadratic form function.**

    Parameters:
        n (int): Dimension of the quadratic form.

    Returns:
        tuple: Tuple containing the symmetric matrix A, vector b, and scalar c.
    """
    eigenvalues = generate_tuple_eigenvalues(n)
    A, _ = symmetric_matrix(eigenvalues)
    b = np.random.rand(n)
    c = np.random.uniform(-50, 50)
    return A, b, c

def random_weakly_h_test(n):
    """
    **Generate a random weakly convex quadratic form function.**

    Parameters:
        n (int): Dimension of the quadratic form.

    Returns:
        function: Random weakly convex quadratic form function.
    """
    A, b, c = generate_random_parameters(n)
    def h_test(x):
        return h(x, A, b, c)
    return h_test

# Example usage (n = 2 to plot):
n = 2
h_test = random_weakly_h_test(n)

# Plot new random weakly convex restriction h_test(x)
fig2 = plt.figure(figsize=(12, 5))
ax1 = fig2.add_subplot(121, projection='3d')
plot_function(ax1, h_test, [-20, 20], [-20, 20], 'Weakly-convex function $h_{test}(x)$')
plt.tight_layout()
plt.show()
```

```python
def generate_nonlinear_setting(n, m):
    """
    **Generate nonlinear constraints for optimization.**

    Parameters:
        n (int): Dimension of the quadratic form.
        m (int): Number of constraints.

    Returns:
        tuple: Tuple containing the combined constraint function and NonlinearConstraint object.
    """
    h_funcs = [random_weakly_h_test(n) for _ in range(m)]
    def h_combined(x):
        return np.array([h(x) for h in h_funcs])
    upper_bounds = np.zeros(len(h_funcs))
    lower_bounds = -np.inf * np.ones(len(h_funcs))
    constraint = NonlinearConstraint(h_combined, lower_bounds, upper_bounds)
    return h_combined, constraint

def find_x_given_multiple_restriction(h_combined, n, m):
    """
    **Find a point satisfying multiple constraints.**

    Parameters:
        h_combined (function): Combined constraint function.
        n (int): Dimension of the quadratic form.
        m (int): Number of constraints.

    Returns:
        numpy.ndarray: Point satisfying the constraints.
    """
    upper_bounds = np.inf * np.ones(m)
    lower_bounds = -np.inf * np.ones(m)
    lower_bounds[0] = 0.5
    constraint = NonlinearConstraint(h_combined, lower_bounds, upper_bounds)
    initial_guess = np.ones(n)
    def objective_function(x):
        return 0
    result = minimize(objective_function, initial_guess, method='trust-constr', constraints=constraint)
    x_given = result.x
    return x_given

def check_intersection_non_empty(constraint, n):
    """
    **Check if the intersection of constraints is non-empty.**

    Parameters:
        constraint (NonlinearConstraint): NonlinearConstraint object.
        n (int): Dimension of the quadratic form.

    Returns:
        bool: True if the intersection is non-empty, False otherwise.
    """
    def objective_function(x):
        return 0
    initial_guess = np.ones(n)
    result = minimize(objective_function, initial_guess, method='trust-constr', constraints=constraint)
    if result.success:
        return True
    else:
        return False
```

#### Test: Counting how many times the solution obtained during the optimization of a feasible problems results a feasible solution

```python
from tqdm import tqdm

def count_feasible_solutions_multiple_restriction(n, m, k):
    """
    **Perform optimization and count feasible solutions among multiple feasible problems.**

    Parameters:
        n (int): Dimension the state space.
        m (int): Dimension of the constraints sets.
        k (int): Number of problems to solve.

    Returns:
        tuple: A tuple containing:
            - The number of feasible problems encountered.
            - The number of infeasible problems encountered.
            - The count of feasible solutions obtained.
    """
    tolerance = 0.000000001
    feasible_problem = 0
    feasible_count = 0
    
    for _ in tqdm(range(k), desc="Progress", unit="Problem"):
        h_combined, constraint = generate_nonlinear_setting(n, m)
        
        if check_intersection_non_empty(constraint):
            feasible_problem += 1
            x_given = find_x_given_multiple_restriction(h_combined, n, m)  
            x_0 = np.random.uniform(-10, 10, n)
            def objective_function(y):
                return objetive_function_aux(y, x_given)
            sol = minimize(objective_function, x_0, method='trust-constr', constraints=constraint)
            if all(i <= tolerance for i in h_combined(sol.x)):
                feasible_count += 1
            else:
                print("Infeasible solution found!: Problem", _, "/", k,
                      "| Solution obtained: x =", sol.x,
                      "| Evaluation h(x) =", h_combined(sol.x))
                
    unfeasible_problem = k - feasible_problem
    return feasible_problem, unfeasible_problem, feasible_count
```

#### Test: Dimension of space: $n = 3$ | Number of restrictions: $m = 2$ | Number of optimization problems to solve: $k = 10$

```python
# Parameters
n = 10  
m = 2
k = 10
feasible_problem, unfeasible_problem, feasible_count = count_feasible_solutions_multiple_restriction(n, m, k)
print("Number of feasible problems:", feasible_problem, "/", k,
      "| Number of feasible solutions beetween feasible problems:",feasible_count, "/", feasible_problem,
      "| Number of unfeasible problems:", unfeasible_problem)
```
